****What are the potential consequences of sampling with versus without replacement?****
* If we were to sample with replacement, there is a chance that individuals will be counted twice, which could skew the mean. When we sample without replacement, each time an individual is selected for the sample and removed from the population, the likelihood of drawing a different individual, increases, and allows us to create a sample of unique individuals that will hopefully represent the population as accurately as possible.

### 2. Markov Chain: physical exercise training method A is used only 5% of the time, a person using method A will stay with this method 85% of the time, and a person not using method A will switch to method A about 65% time. At the beginning of the experiment only 5% of people used method A.

#### 1. Generate a transition matrix for this Markov chain
#### 2. Generate a transition plot (using R or by hand as an image it’s valid)
#### 3. Plot the change in the probabilities over time for both methods until the 10th time unit.

* What are the key properties of a transition matrix in a Markov chain?
* What does it mean for a Markov chain to reach a steady-state distribution? When is this achieved in this analysis?


```r
#import package:
library("markovchain")
```

#### 1. Generate a transition matrix for this Markov chain

```r
#define different states
methodStates <- c("A", "B")

#set byRow as True
byRow <- TRUE

#create transitional matrix for exercise program
exerciseMatrix <- matrix(data = c(0.85, 0.15, #list out transition states for A and B
                                  0.65, 0.35), byrow = byRow, nrow = 2,
                        dimnames = list(methodStates, methodStates)) 

#make the matrix a markovchain class
mcExercise <- new("markovchain", states = methodStates, byrow = byRow,
                 transitionMatrix = exerciseMatrix, name = "Exercise Method")
print(mcExercise)
```

The transition matrix for this analysis is:

|   |   A  | B    |
|---|------|------|
| A | 0.85 | 0.15 |
| B | 0.65 | 0.35 |

#### 2. Generate a transition plot (using R or by hand as an image it’s valid)

```r
#create plot for mcExercise
options(repr.plot.width=11, repr.plot.height=8) #adjust plot dimensions (was being cut off)
plot(mcExercise, edge.curved = 0.1) #plot transition state
```

#### 3. Plot the change in the probabilities over time for both methods until the 10th time unit.

```r
#plot the change in the probabilities for both exercise methods until the 10th time unit

#define the initial state
initialState <- c(0.05, 0.95) #There is a 5% chance that someone will be in method A at the start

#view what the distribution is after iterating 10 times:
mcExercise10 <- initialState * (mcExercise ^10)
mcExercise10
```

#### What are the key properties of a transition matrix in a Markov chain? What does it mean for a Markov chain to reach a steady-state distribution? When is this achieved in this analysis?

```r
#calculate the steady-state distribution 
steady_state <- steadyStates(object = mcExercise)
print(steady_state)
```

The key properties of a transition matrix in a Markcov chain are the initial states and the changing states. In this analysis, the initial states are the probability of being in Method A (5%) or Method B (95%) at the beginning of the experiment. The change states refer to the probability of changing from one state to another. When someone is in Method A, there is an 85% probability that they will stay there and a 15% probability that they will change to Method B. Alternatively, for someone who is in Method B, the probability that they will change to Method A is 65% and the probability that they will stay in Method B is 35%. 

The steady state distribution refers to the longterm probability of being in either of the states. In this analysis, steady-state is achieved after X number of iterations. The steady-state distribution for this analysis is A: 81.25% and B: 18.75%, so that once steady-state is reached, the probability of being in Method A is 81.25% and the probability of being in Method B is 18.75%.

#### 3. Random Walk: Another simpler example of a random walk is a one-dimensional random walk. first we place a marker at zero (our initial state), we flip a coin, if it lands on heads, the marker is moved one unit to the right (1), if it lands on tails it is moved one unit to the left.

#### 1. Generate a function that randomly draws from our initial state and populates a vector with the different transitions.

#### 2. Generate a plot that shows 500 independent one-dimensional walks, differentiating walks that end above 0 or below 0.

#### 3. What is the frequency of walks that ended in a positive cumulative count, in zero, or negative?

* How does a random walk differ from a Markov process?

****1. Generate a function that randomly draws from our initial state and populates a vector with the different transitions.****

```r
#generate a random walk function 
random_walk = function(N){ #N steps
    
    #define initial position
    position <- 0 
    walk_vect <- numeric(N) #create vector for results
    
    for (i in 1:N) { #for range of 1,N number of steps:
        
        walk_val <- sample(c(-1,1), size = 1) #pick a value (1 or -1)
        position <- position + walk_val #add to current position
        walk_vect[i] <- position #add to vector
    }   
    return(walk_vect)
}
```

****2. Generate a plot that shows 500 independent one-dimensional walks, differentiating walks that end above 0 or below 0.¶****

```r
#create a data frame from the function and plot dataframe:

#plotting 500 independent walks
N= 500

#create data frame where X = which value in 1:N, and Y = the resulting walk position
random_walk_df = data.frame(
    walk = 1:N,
    position = random_walk(N)
)

#add color to values that are above/below 0
random_walk_df$color <- ifelse(random_walk_df$position > 0, "Above 0",
                               ifelse(random_walk_df$position < 0, "Below 0", "0"))
#head(random_walk_df)

#plot step number against walk position
ggplot(random_walk_df,aes(x = walk, y = position)) + #x = i, y = resulting position
geom_line() + 
geom_point(aes(color = color), size=1)+ #use color to indicate above/below zero
scale_color_manual(values = c("Above 0" = "lightblue", "Below 0" = "red", "0" = "grey")) +
ggtitle("One-dimensional walk with 500 steps")
```

****3. What is the frequency of walks that ended in a positive cumulative count, in zero, or negative?****

```r
#calculate frequencies:

#get counts for each category in the random walk
category_counts <- random_walk_df %>%
    count(color)

#make counts a factor to use for scale
category_counts$color <- factor(category_counts$color, levels = c("Above 0", "Below 0", "0"))

ggplot(category_counts, aes(x=color, y = n, fill = color)) +
    geom_bar(stat = "identity") + 
    scale_fill_manual(values = c("Above 0" = "lightblue", "Below 0" = "red", "0" = "grey")) +
    ggtitle("Frequency of random walk positions")
```

When running the random walk multiple times, these frequencies greatly varied. Occassionally, the distribution of points above and below zero was relatively even, but often it would randomly trend in one direction.

****How does a random walk differ from a Markov process?****

One of the key differences between a random walk and a Markov process is that changes in a random walk happen independently of previous states and are only dependent on the current state, whereas change states in a Markov process are influenced by previous states. This is demonstrated when we rerun the random walk and do not see the same kind of steady-state that we do with the Markov chain.
